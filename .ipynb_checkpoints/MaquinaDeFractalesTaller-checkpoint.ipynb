```python
# Importamos todas las librerías necesarias
import numpy as np #manejo de números, matrices, vectores
import random #con esto introducimos el uso de la aleatoriedad
import matplotlib.pyplot as plt #con esto podemos graficar



#La siguiente es la función que aplica el teorema de punto fijo

def PuntoFijo(Transformaciones,cantIter,vectorInicial):
  Fractal = [vectorInicial]
  cantTransf = len(Transformaciones[0])
  for i in range(1,cantIter+1):
    eleccion = random.randrange(0,cantTransf)
    nuevoPunto = Transformaciones[0][eleccion]@Fractal[i-1] + Transformaciones[1][eleccion]
    Fractal = Fractal + [nuevoPunto]

  return(Fractal)

'''
Acá se definene funciones que devuelven algunas transformaciones usuales 
dependiendo de ciertos parámetros
'''

# la siguiente función devuelve la matriz de una rotación en R^2 de angulo ang
def rotacion(ang):
  R = np.array([[np.cos(ang),np.sin(ang)],[-np.sin(ang),np.cos(ang)]])
  return(R)

'''
 la siguiente función devuelve una lista de rotaciones y traslaciones al azar
 tantas de cada una cómo el parámetro 'cantTrans'
'''
def RotacionesTraslacionesAlAzar(cantTrans):
  TransfLineal = [rotacion(np.pi*np.random.rand()) for i  in range(0,cantTrans)]
  Traslacion = [np.random.rand(2,1) for i  in range(0,cantTrans)]

  return([TransfLineal,Traslacion])

'''
 la siguiente función devuelve una matriz que contrae/dilata dependiendo de 
 los parámetros 'a' y 'b' en cada dirección canónica del plano
'''
def Diagonal(a,b):
  D = [[a,0],[0,b]]
  return(D)

'''
 la siguiente función devuelve una lista de diagonales y traslaciones al azar
 tantas de cada una cómo el parámetro 'cantTrans'
'''
def DiagonalesTraslacionesAlAzar(cantTrans):
  TransfLineal = [ Diagonal(np.random.rand(),np.random.rand()) for i  in range(0,cantTrans)]
  Traslacion = [np.random.rand(2,1) for i  in range(0,cantTrans)]

  return([TransfLineal,Traslacion])

'''
Las siguiente transformaciones son las del triangulo de sierpinsky 
y una generalización en la que se contrae dependiendo de un parámetro toda la figura.
'''

def TransfSierpinski():
  T = (1/2)*np.identity(2)
  TransfLineal = [T,T,T]
  Traslacion = [[0,0],[0,1/2],[1/4,1/2]]
  return([TransfLineal,Traslacion])

def TransfSierpinskiContraccion(contraccion):
  T = contraccion*np.identity(2)
  TransfLineal = [T,T,T]
  Traslacion = [[0,0],[0,1/2],[1/4,1/2]]
  return([TransfLineal,Traslacion])

def DibujarSierpinski(contraccion,cantIter):
  Transformaciones = TransfSierpinskiContraccion(contraccion)
  vectorInicial = [0,0]
  PF = PuntoFijo(Transformaciones,cantIter,vectorInicial)
  x = [PF[i][1] for i in range(1,cantIter+1)]
  y = [PF[i][0] for i in range(1,cantIter+1)]
  plt.plot(x,y,'.')
  plt.show()

def DibujarMuchoSierpinski(CantidadSierpinski):
    ListaSierpinskis = []
    for s in range(1,CantidadSierpinski):
        ListaSierpinskis = ListaSierpinskis + [DibujarSierpinski(s/CantidadSierpinski,10000)]

'''    
A continuación definimos las funciones que permiten dibujar 
una versión del triangulo de Sierpinski que en cada transformación contrae por un parámetro distinto
'''
def TransfSierpinskiContraccionES(contraccion_1, contraccion_2, contraccion_3):
  T = np.identity(2)
  TransfLineal = [contraccion_1*T,contraccion_2*T,contraccion_3*T]
  Traslacion = [[0,0],[0,1/2],[1/4,1/2]]
  return([TransfLineal,Traslacion])

def DibujarSierpinskiContraccionES(contraccion_1, contraccion_2, contraccion_3):
    Transformaciones = TransfSierpinskiContraccionES(contraccion_1, contraccion_2, contraccion_3)
    cantIter = 40000
    vectorInicial = [0,0]
    PF = PuntoFijo(Transformaciones,cantIter,vectorInicial)
    x = [PF[i][1] for i in range(1,cantIter+1)]
    y = [PF[i][0] for i in range(1,cantIter+1)]
    plt.plot(x,y,'.')
    plt.show()

def DibujarSierpinskiContraccionESGIF(contraccion_1, contraccion_2, contraccion_3):
    Transformaciones = TransfSierpinskiContraccionES(contraccion_1, contraccion_2, contraccion_3)
    cantIter = 40000
    vectorInicial = [0,0]
    PF = PuntoFijo(Transformaciones,cantIter,vectorInicial)
    x = [PF[i][1] for i in range(1,cantIter+1)]
    y = [PF[i][0] for i in range(1,cantIter+1)]
    plt.plot(x,y,'.')
    plt.savefig('SierpinskiContracciones-'+str(contraccion_1)+'-'+str(contraccion_2)+'-'+str(contraccion_3)+'.png')#, bbox_inches='tight')
    plt.clf()


'''    
A continuación definimos las funciones que permiten dibujar 
una versión del triangulo de Sierpinski que en cada transformación rota por un parámetro distinto
'''

def TransfSierpinskiRotacionES(ang_1, ang_2, ang_3):
  TransfLineal = [0.5*rotacion(ang_1),0.5*rotacion(ang_2),0.5*rotacion(ang_3)]
  Traslacion = [[0,0],[0,1/2],[1/4,1/2]]
  return([TransfLineal,Traslacion])


def DibujarSierpinskiRotacionES(ang_1, ang_2, ang_3):
    Transformaciones = TransfSierpinskiRotacionES(ang_1, ang_2, ang_3)
    cantIter = 40000
    vectorInicial = [0,0]
    PF = PuntoFijo(Transformaciones,cantIter,vectorInicial)
    x = [PF[i][1] for i in range(1,cantIter+1)]
    y = [PF[i][0] for i in range(1,cantIter+1)]
    plt.plot(x,y,'.')
    plt.show()

def DibujarSierpinskiRotacionESGIF(ang_1, ang_2, ang_3):
    Transformaciones = TransfSierpinskiRotacionES(ang_1, ang_2, ang_3)
    cantIter = 40000
    vectorInicial = [0,0]
    PF = PuntoFijo(Transformaciones,cantIter,vectorInicial)
    x = [PF[i][1] for i in range(1,cantIter+1)]
    y = [PF[i][0] for i in range(1,cantIter+1)]
    plt.plot(x,y,'.')
    plt.savefig('SierpinskiRotaciones-'+str(ang_1)+'-'+str(ang_2)+'-'+str(ang_3)+'.png')#, bbox_inches='tight')
    plt.clf()

```

```python
'''
En esta celda de código se corren los programas que generan un archivo .gif 
En la animación se muestra la evolución de un fractal que se va deformando cambiando las transformaciones a través de un parámetro.
En este caso el parámetro que se modifica es un ángulo de rotación.
Es muy interesante notar que todos los fractales que se muestran en esta animación tienen la misma dimensión que el triángulo de Sierpinski clásico.
'''
AnguloFinal = 2*np.pi
Cantidad = 80
for n in range(Cantidad):
    DibujarSierpinskiRotacionESGIF((n/Cantidad)*AnguloFinal,-(n/Cantidad)*AnguloFinal,0)


import os
#import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

image_dir = os.getcwd()
image_files = sorted([os.path.join(image_dir, file) for file in os.listdir(image_dir) if file.endswith(".png")])

# Function to update the animation
def update(frame):
    plt.clf()  # Clear the current plot
    plt.imshow(plt.imread(image_files[frame]))  # Display the next image
    plt.axis('off')  # Turn off axis
    plt.title(f"Frame {frame}")  # Set title to current frame number


# Create a figure and axis
fig, ax = plt.subplots()

# Create the animation
animation = FuncAnimation(fig, update, frames=len(image_files), interval=200)

# Save the animation as a gif
animation.save('animation.gif', writer='imagemagick', fps=30)

# Show the animation
plt.show()
```

```python
'''
Acá definimos algunas otras transformaciones que definen otras familias de fractales, algunas mas conocidas y otros hallazgos fortuitos
'''


# Esta familia de fractales tiene cierta reminicencia con los árboles
def TranfArbol(contraccion):
  #acá se deciden 4 matrices concretas a hardcodear
  A = np.array(contraccion*rotacion(np.pi/4))
  B = np.array(contraccion*rotacion(-np.pi/4))
  C = np.array(np.identity(2))
  TransfLineal = [A,B,C]
  #acá se deciden 4 puntos concretos a harcodear para trasladar (o sea sumar) por ellos
  ValueError
  v_A = [1,0]
  v_B = [0,1]
  v_C = [0,0]
  Traslacion = [v_A,v_B,v_C]
  return([TransfLineal,Traslacion])

def DibujarArbol(contraccion,cantIter):
  Transformaciones = TranfArbol(contraccion)
  vectorInicial = [0,0]
  PF = PuntoFijo(Transformaciones,cantIter,vectorInicial)
  x = [PF[i][1] for i in range(1,cantIter+1)]
  y = [PF[i][0] for i in range(1,cantIter+1)]
  plt.plot(x,y,'o')
  plt.show()

```

```python
# cambiando el parametro de contraccion conseguimos diferentes fractales
contraccion = 0.7

DibujarArbol(contraccion,10000)
```

```python
'''
Los diferentes fractales que se dibujan en esta familia tienen diferentes dimensiones
'''

# la siguiente funcion sirve para dibujar muchos arboles cambiando el parametro de contraccion

def DibujarMuchosArboles(contraccion_inicial, contraccion_final, cantidad):
  c_0 = contraccion_inicial
  c_f = contraccion_final
  for k in range(1,cantidad):
    contraccion = (1-k/cantidad)*c_0 + (k/cantidad)*c_f
    DibujarArbol(contraccion,20000)

# cambiando los siguiente parametros podemos obtenes distintas familias de fractales

contraccion_inicial = 0.3
contraccion_final = 0.6
cantidad = 20

DibujarMuchosArboles(contraccion_inicial,contraccion_final,cantidad)
```

```python
'''
Una última familia de transformaciones que induce otra familia de fractales
'''

def TransfOtro(contraccion):
  #acá se deciden 4 matrices concretas a hardcodear
  A = np.array(contraccion*np.identity(2))
  B = np.array((1/3)*rotacion(np.pi/2))
  C = np.array(contraccion*np.identity(2)@Diagonal(1,-1/2))
  D = contraccion * np.array(rotacion(np.pi/4)@ Diagonal(3/4,-1/2) @ rotacion(np.pi/3))
  TransfLineal = [A,B,C,D]
  #acá se deciden 4 puntos concretos a harcodear para trasladar (o sea sumar) por ellos
  v_A = [1,0] #[1/4,1/2]
  v_B = [1/2,-1/4] #[0,1/3]
  v_C = [1/4,1/2] #[1/2,-1/4]
  v_D = [0,1/3] #[1,0]
  Traslacion = [v_A,v_B,v_C,v_D]
  return([TransfLineal,Traslacion])

def DibujarOtro(contraccion,cantIter):
  Transformaciones = TransfOtro(contraccion)
  vectorInicial = [0,0]
  PF = PuntoFijo(Transformaciones,cantIter,vectorInicial)
  x = [PF[i][1] for i in range(1,cantIter+1)]
  y = [PF[i][0] for i in range(1,cantIter+1)]
  plt.plot(x,y,'.')
  plt.show()


```

```python
#modificando el parámetro de contracción podemos conseguir diferentes fractales, cada uno de ellos con dimensión diferente

contraccion = 29/51

DibujarOtro(contraccion,10000)
```

```python
'''
Los diferentes fractales que se dibujan en esta familia tienen diferentes dimensiones
'''

def DibujarMuchosOtros(contraccion_inicial, contraccion_final, cantidad):
  c_0 = contraccion_inicial
  c_f = contraccion_final
  for k in range(1,cantidad):
    contraccion = (1-k/cantidad)*c_0 + (k/cantidad)*c_f
    DibujarOtro(contraccion,10000)
```
